Cz 1. **Definicja podanej w zadaniu 3 funkcji**


```python
def custom_function(x):
    return x**2+5
```

Cz 2. **Wywołanie funkcji dla podanych w zadnaiu 3 zakresów dziedziny**

```python
import numpy as np

x1 = np.linspace(-1,1,200)
x2 = np.linspace(-6,6,1200)
x3 = np.linspace(0,5,500)

y1 = custom_function(x1)
y2 = custom_function(x2)
y3 = custom_function(x3)
```

Cz 3. **Wizualizacja wyników poprzez wykresy**

```python
import matplotlib.pyplot as plt

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)

plt.legend(["x>-1 i x<1", "x>-6 i x<6", "x>0 i x<5"])
plt.title("Wizualizacja otrzymanych przebiegów - wspolny wykres")
plt.xlabel("x")
plt.ylabel("custom_function(x)")

plt.show()

plt.subplot(1, 3, 1)
plt.plot(x1,y1)
plt.title("x>-1 i x<1")
plt.ylabel("custom_function(x)")
plt.xlabel("x")

plt.subplot(1, 3, 2)
plt.plot(x2,y2)
plt.title("x>-6 i x<6")
plt.xlabel("x")

plt.subplot(1, 3, 3)
plt.plot(x3,y3)
plt.title("x>0 i x<5")
plt.xlabel("x")

plt.suptitle("Wizualizacja otrzymanych przebiegów - osobne wykresy")
plt.show()
```
